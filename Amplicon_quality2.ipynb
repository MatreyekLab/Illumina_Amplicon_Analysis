```python
list_positions = [90,111,115,122,102,131,142,146,166,168,175]
list_AA = []

import pandas as pd
df = pd.read_csv ('Output_Amplicon.csv')
df.head()
```

```python
newdf = df.groupby(['Position','AminoAcids','Quality']).Quality.agg('count').to_frame('Count').reset_index()
newdf.head(50)
```

```python
table = pd.pivot_table(newdf, values=['Count'], 
                       index=['Position','AminoAcids'], columns=['Quality']).fillna(0.0)
table.head(50)
```

```python
table.to_csv('Amplicon_Quality.csv')
```
